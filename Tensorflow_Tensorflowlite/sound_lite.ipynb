```python
RESULT_PATH = './results/'

model_keras= tf.keras.models.load_model(RESULT_PATH+"/model.h5")
# Converting a tf.Keras model to a TensorFlow Lite model.
converter = tf.lite.TFLiteConverter.from_keras_model(model_keras)
tflite_model_from_keras_model = converter.convert()

with open(RESULT_PATH+'model_from_keras_model.tflite', 'wb') as f:
  f.write(tflite_model_from_keras_model)
```
